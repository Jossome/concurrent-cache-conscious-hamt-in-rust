```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('dump/vs_get_1000.csv')
df['time_hash'] = df['time_hash'][0] / df["time_hash"]
df['time_cctrie'] = df['time_cctrie'][0] / df['time_cctrie']
_, ax = plt.subplots()
ax.plot(df['n_thread'], df['time_hash'])
ax.plot(df['n_thread'], df['time_cctrie'])
ax.set_title("1000 nodes, n_threads = 1,2,4,8")
ax.set_xlabel('nthreads')
ax.set_ylabel("Speed up percentage")
ax.legend(["hash", "cctrie"])
```
