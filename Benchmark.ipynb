```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('dump/vs_get_1000.csv')

f = plt.figure(figsize=(12,4))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)

ax.plot(df['n_thread'], df['time_hash'])
ax.plot(df['n_thread'], df['time_cctrie'])
ax.set_title("1000 nodes execution time")
ax.set_xlabel('nthreads')
ax.set_ylabel("time")
ax.legend(["hash", "cctrie"])

df['time_hash'] = df['time_hash'][0] / df["time_hash"]
df['time_cctrie'] = df['time_cctrie'][0] / df['time_cctrie']
ax2.plot(df['n_thread'], df['time_hash'])
ax2.plot(df['n_thread'], df['time_cctrie'])
ax2.set_title("10000 nodes speed up")
ax2.set_xlabel('nthreads')
ax2.set_ylabel("Speed up percentage")
ax2.legend(["hash", "cctrie"])
```
