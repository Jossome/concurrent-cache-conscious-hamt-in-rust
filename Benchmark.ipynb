```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('dump/24_8_compare.csv', sep=' ', names=['10e', 'amount', 'speedup'], index_col='10e')
_, ax = plt.subplots()
ax.plot(df['speedup'])
ax.set_title('key length=24, key_group=8, best-case')
ax.set_xlabel('10 power by')
ax.set_ylabel("Speed up percentage")
ax.axhline(y=1, color='r', linestyle='--', lw=2)
```
